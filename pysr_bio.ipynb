```python
import pysr
import sympy
import numpy as np
from matplotlib import pyplot as plt
from pysr import PySRRegressor
import csv
import pandas as pd
data = pd.read_csv("bio_data.csv")
function_set = ['add', 'mul', 'sub']
```

**ds1/dt**

```python
# Separate the data into the 'Ii' and 'f3' variables
data = pd.read_csv("bio_data_234.csv")
X = data[['X1','X6']].values
y = data['f_pred_1'].values


pysr_params = dict(
    populations=30,
    model_selection="best",
)


model = PySRRegressor(
    niterations=30,
    binary_operators=["plus", "sub", "mult", "div"],
    # Include any unary operators if you think they might be helpful
    **pysr_params
)

# Run model:

model.fit(X, y)


print(f"Model: {model}")

print(f"dS1/dt =  {model.sympy()}")

#model.sympy(2)
print(f"dS1/dt =  {model.latex()}")
```

**dS2/dt**

```python

X = data[['X1','X2','X5', 'X6']].values
y = data['f_pred_2'].values


# PySR parameters
pysr_params = dict(
    populations=1000,
    model_selection="best",
)

# Create and configure the PySRRegressor model
model = PySRRegressor(
    niterations=30,
    binary_operators=["+", "-", "*", "/"],
    ncyclesperiteration=10,
    parsimony=0.001,  # Increasing this value increases the penalty for complexity
    warm_start=True,
    **pysr_params
)


# model = PySRRegressor(
#     niterations=100,
#     binary_operators=["plus", "mult", "sub", "div", "pow"],
#     maxsize=20,
#     maxdepth=10,
#     parsimony=0.001,
#     alpha=0.001,
#     **pysr_params
# )

# model = PySRRegressor(
#     niterations=30,
#     binary_operators=["plus", "mult", "sub"],
#     maxsize=20,
#     **pysr_params
# )

# model = PySRRegressor(
#     niterations=30,
#     binary_operators=["plus", "mult", "sub", "pow"],
#     maxsize=20,
#     maxdepth=10,
#     parsimony=0.001,
#     alpha=0.001,
#     **pysr_params
# )

model.fit(X, y)

print(f"Model: {model}")

print(f"dS2/dt= {model.sympy()}")

#model.sympy(2)
print(f"dS2/dt= {model.latex()}")
```
