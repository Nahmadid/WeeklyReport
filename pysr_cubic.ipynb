```python

import pysr

import sympy
import numpy as np
from matplotlib import pyplot as plt
from pysr import PySRRegressor
import csv

import pandas as pd
data = pd.read_csv("combined_data.csv")
function_set = ['add', 'mul', 'sub']
# Separate the data into the 'Ii' and 'f3' variables
X = data[['X1', 'X2']].values
y = data['f_pred_2'].values
```

```python




pysr_params = dict(
    populations=30,
    model_selection="best",
)


model = PySRRegressor(
    niterations=30,
    binary_operators=["plus", "mult"],
    **pysr_params
)

# Run model:

model.fit(X, y)


print(f"Model: {model}")

print(f"Model SymPy: {model.sympy()}")

#model.sympy(2)
print(f"Latex Equation: {model.latex()}")
```
